```python
candidate = project_root / 'data' / 'synthetic' / 'signup_experiment.csv'
if not candidate.exists():
    raise FileNotFoundError('Run `make demo` first so the synthetic case-study data exists.')

df = pd.read_csv(candidate)
df.head()
```

```python
analysis = analyze_experiment(
    input_path=str(candidate),
    metric_column='outcome',
    metric_type='binary',
    covariate_column='pre_metric',
    sequential_looks=5,
    guardrail_columns=['support_tickets'],
    guardrail_directions={'support_tickets': 'lower_is_better'},
    segment_column='segment',
)
analysis['result']
```
